```python
! pip install selenium
from selenium import webdriver
from time import sleep
from selenium.webdriver.common.by import By
import pandas as pd
from selenium.webdriver.common.keys import Keys
```

## Set Path for Driver and Open the Browser 

```python
path="./selenium_driver/"
driver=webdriver.Chrome(executable_path=path)
```

## Load the WebPage

```python
driver.get("https://www.amazon.com")
driver.maximize_window()
```

## Get Input Elements

```python
input_search=driver.find_element(By.ID,"twotabsearchtextbox")
search_button=driver.find_element(By.XPATH,"(//input[@class='nav-input nav-progressive-attribute'])[2]")
```

## Send Input to the WebPage

```python

input_search.send_keys("Mobile Phones")
sleep(0.1)
search_button.click()
```

## Now Start Scraping 

```python
Products = []
Price = []


for i in range(2):
    print(f"Scraping page {i+1}")
    product = driver.find_elements(By.XPATH, "//*[contains(@class,'a-size-base-plus a-color-base a-text-normal') or contains(@class,'a-size-medium a-color-base a-text-normal')]")
    price = driver.find_elements(By.XPATH, "//span[@class='a-price-whole']")

    
    for p in product:
        Products.append(p.text)
    for s in price:
        Price.append(s.text)
    
```

```python

print(len(Products))
print(len(Price))
```

```python

df = pd.DataFrame({"Product Names": Products,"Price($)":Price})
df
```

```python
df.to_csv("./Mobile Phone Names and prices.csv")
```
