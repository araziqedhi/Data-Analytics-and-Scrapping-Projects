## Creating Match Results File

```python
# Connect to Website and pull in data

URL = "https://www.espncricinfo.com/series/indian-premier-league-2023-1345038/match-schedule-fixtures-and-results"
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/78.0.3904.108 Safari/537.36", "Accept-Encoding":"gzip, deflate", "Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8", "DNT":"1","Connection":"close", "Upgrade-Insecure-Requests":"1"}

page = requests.get(URL, headers=headers)

soup1 = BeautifulSoup(page.content, "html.parser")

soup2 = BeautifulSoup(soup1.prettify(), "html.parser")
soup2
```

```python
soup2.title
```

```python
data= []
ground_elements = soup2.find_all(class_="ds-text-tight-xs ds-truncate ds-text-typo-mid3")
for ground_element in ground_elements:
    ground_name = ground_element.text.strip()
    data.append(ground_name)
```

```python
data
```

```python
import re



# Regular expression patterns for extracting match number, date, and ground name
pattern_match_no = r'(\d+)[a-z]+\s'
pattern_match_no = r'(?:Qualifier|Eliminator|Final)?\s?(\d+)[a-z]*\s'

pattern_date = r'(\w+\s\d{1,2}( - \d{1,2})?,\s\d{4})'
pattern_ground = r',\s(.*?),'

match_numbers = []
dates = []
ground_names = []

for item in data:
    match_no = re.search(pattern_match_no, item)
    date = re.search(pattern_date, item)
    ground = re.search(pattern_ground, item)

    if match_no is not None:
        match_numbers.append(match_no.group(1))
    if date is not None:
        dates.append(date.group(1))
    if ground is not None:
        ground_names.append(ground.group(1))

for i in range(len(dates)):
    print("Match No:", match_numbers[i])
    print("Date:", dates[i])
    print("Ground Name:", ground_names[i])
    print("---------")
```

```python
print(len(dates))
print(len(ground_names))
```

```python
element=soup2.find_all(class_="ds-flex ds-items-center ds-min-w-0 ds-mr-1", attrs={"title": True})
team1=[]
for i in element:
    title=i["title"]
    print(title)
    team1.append(i)
print(team1)
```

```python

element = soup2.find_all(class_="ds-flex ds-items-center ds-min-w-0 ds-mr-1", attrs={"title": True})
team1 = []

for i in element:
    title = i["title"]
    team1.append(title)

print(team1)


```

```python
print(len(team1))
team1=team1[12:]
team1part2=team1[0::2]
print(len(team1part2))
team2=team1[1::2]
print(len(team1))
print(len(team2))
print(team1)
print(team2)
```

```python
spans = soup2.find_all(class_="ds-text-tight-s ds-font-regular ds-line-clamp-2 ds-text-typo")

team_names = []
margins = []

for span in spans:
    text = span.get_text(strip=True)
    if " won by " in text:
        parts = text.split(" won by ")
        team_name = parts[0]
        margin = parts[1]
        team_names.append(team_name)
        margins.append(margin)
    elif "No result" in text:
        team_name="Drawn"
        team_names.append(team_name)
        margins.append(team_name)
print("Team Names:", team_names)
print("Margins:", margins)
print(len(team_names))
print(len(margins))
```

```python
print(len(match_numbers))
print(len(dates))
print(len(ground_names))
print(len(team1part2))
print(len(team2))
print(len(team_names))
print(len(margins))
```

```python
filename = 'match_data.csv'

with open(filename, 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Match No', 'Date', 'Ground Name',"Team1","Team2","Winning Team","Margin"])  # Write header row
    for i in range(len(match_numbers)):
        writer.writerow([match_numbers[i], dates[i], ground_names[i],team1part2[i],team2[i],team_names[i],margins[i]])
```
